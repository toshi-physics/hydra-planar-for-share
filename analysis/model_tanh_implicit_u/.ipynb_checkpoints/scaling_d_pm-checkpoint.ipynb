```python
import numpy as np
from matplotlib import pyplot as plt
import json, os
from skimage.measure import label, regionprops
from scipy.special import kn
```

```python
run = 1
Ks = np.array([1])
Ds = np.array([1.0, 2.0, 3.0, 6.0, 10.0])
betas = np.array([[1.5, 2.0, 2.5, 3.0], [3.0, 4.0, 5.0, 6.0], [4.5, 6.0, 7.5, 9.0], [9, 12.0, 15.0, 18.0], [11.0, 13.0, 15.0, 17.0]])
alpha = 0.1
chi = 0.0
gamma = 5.0

ddir = 'data' #'/new_drho_data'#
model = '/model_tanh_implicit_no_friction/' # '/model_hill'#

pdir = '../../'+ddir+model+'/D_{:.1f}_K_{:d}/alpha_{:1.1f}_beta_{:1.4f}_chi_{:1.1f}_gamma_{:1.1f}/run_{:d}/'.format(Ds[0], Ks[0], alpha, betas[0,0], chi, gamma,  run)

datadir = pdir+'data/'

with open(pdir+'parameters.json') as jsonFile:
    parameters = json.load(jsonFile)

T         = parameters["T"]        # final time
dt_dump   = parameters["dt_dump"]
n_steps   = int(parameters["n_steps"])  # number of time steps
n_dump = int(T/dt_dump); dn_dump = int(n_steps/n_dump)
K         = parameters["K"]        # elastic constant, sets diffusion lengthscale of S with Gamma0
mx        = np.int32(parameters["mx"])
my        = np.int32(parameters["my"])
mu        = np.float32(parameters["mu"])
dx        = np.float32(parameters["dx"])
dy        = np.float32(parameters["dy"])
tauc      = np.float32(parameters["tauc"])
#setup a meshgrid
tol = 0.001

x   = np.linspace(0+tol, mx*dx-tol, mx)
y   = np.linspace(0+tol, mx*dx-tol, my)
xv, yv  = np.meshgrid(x,y, indexing='ij')


times = np.arange(0, n_dump, 1)*dt_dump

print(n_dump, T, mx, dx, tauc)
```

```python
defect_loc = np.zeros((len(Ds), len(betas[0]), 4, 2))
zero_loc = np.zeros((len(Ds), len(betas[0])))
for di, D in enumerate(Ds):
    for ib, b in enumerate(betas[di]):
        n=19
        pdir = '../../'+ddir+model+'/D_{:.1f}_K_{:d}/alpha_{:1.1f}_beta_{:1.4f}_chi_{:1.1f}_gamma_{:1.1f}/run_{:d}/'.format(D, Ks[0], alpha, b, chi, gamma,  run)
        Qxx = np.loadtxt(pdir+'data/Qxx.csv.{:d}'.format(n), delimiter=',')
        Qxy = np.loadtxt(pdir+'data/Qxy.csv.{:d}'.format(n), delimiter=',')
        eiso = np.loadtxt(pdir+'data/eiso.csv.{:d}'.format(n), delimiter=',')

        zero_loc[di, ib] = np.where(eiso[:, mx//2]>0)[0][0]
        
        S = 2*np.sqrt(Qxx**2+Qxy**2)
    
        mask = S<0.5
        l_mask = label(mask)
        regions = regionprops(l_mask)
        i=0
        for region in regions:
            minx, miny, maxx, maxy = region.bbox
            defect_loc[di, ib, i] = np.array([(minx + maxx )/2,(miny + maxy)/2])
            i+=1
```

```python
fig, axs = plt.subplots(figsize=(5, 5))

dpm_1 = defect_loc[:,:,1, 1] - defect_loc[:,:, 3, 1]
ax = axs
#print(defect_loc)
for i, D in enumerate(Ds):
    d = dpm_1[i] * dx
    lambd = np.sqrt(D)*np.sqrt(tauc)
    b = betas[i]
    nu = (mx//2-zero_loc[i])*dx/lambd
    
    xplot = Ks[0]*lambd**4/(b**2 * nu**2)

    xval = np.log(xplot)
    yval = np.log(d/lambd)

    #p0, p1 = np.polyfit(xval, yval, 1)

    #print(p0, p1)

    #plt.plot(xplot, (xplot**(p0)) * np.exp(p1))
    
    ax.errorbar(xplot, d/lambd, yerr=2*dx/lambd, marker='o', label='D={:.0f}'.format(D), linestyle='')

xx= np.array([3e3, 1e4, 1e5])
plt.plot(xx, 8/xx**(1/4), linestyle='dashed', color='k')

#xx= np.array([5e-9, 1e-8, 1e-7, 1e-6, 2e-6])
#plt.plot(xx, 0.013/xx**(1/4), linestyle='dashed', color='k')

ax.set_xlabel(r'$\dfrac{{K ~\lambda^4}}{{\beta^2 r_0^2}}$', size=14)
ax.set_ylabel(r'$d_{{+-}}/~\lambda$ ', size=14)
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(frameon=False)

#plt.savefig('d_pm_scaling_no_friction.svg', bbox_inches='tight')
#plt.savefig('d_pm_scaling_no_friction.png', bbox_inches='tight')
```

```python
fig, axs = plt.subplots(figsize=(5, 5))

dpm_1 = defect_loc[:,:,1, 1] - defect_loc[:,:, 3, 1]
ax = axs

#print(defect_loc)
for i, D in enumerate(Ds):
    d = dpm_1[i] * dx
    lambd = np.sqrt(D*tauc)
    b = betas[i]
    nu = (mx//2-zero_loc[i])*dx/lambd
    M=2.5
    epsilon_t=0.5
    ct=0.1
    S_bar = alpha*b*tauc*nu**2/(4*epsilon_t*M*lambd**2)
    gamma_net = gamma/(ct**2 + (S_bar/lambd)**2)
    
    xplot = 3*Ks[0]/(gamma_net*S_bar**2)

    yplot = (d/lambd) * (87.4*(lambd/d)**5+2.36*np.exp(-d/lambd))
    
    ax.plot(xplot, yplot, marker='o', label='D={:.0f}'.format(D), linestyle='')

xx=np.array([10, 100, 1000, 10000])
plt.plot(xx, xx, linestyle='dashed', color='k')

ax.set_xlabel(r'$\dfrac{{3K}}{{\gamma \bar{{S}}^2}}$', size=14)
ax.set_ylabel(r'$y(d_{{+-}}/~\lambda)$ ', size=14)
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(frameon=False)

#plt.savefig('d_pm_scaling_no_friction_full_eq.svg', bbox_inches='tight')
#plt.savefig('d_pm_scaling_no_friction_full_eq.png', bbox_inches='tight')
```
