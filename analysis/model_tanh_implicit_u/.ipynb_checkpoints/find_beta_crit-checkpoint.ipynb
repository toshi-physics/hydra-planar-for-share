```python
import numpy as np
from matplotlib import pyplot as plt
import json, os
from skimage.measure import label, regionprops
```

```python
run = 1
K= 1
Ds = [0.1, 0.3, 0.6, 1, 2, 3, 6, 10, 30, 60]
betas = [np.arange(0.1, 0.6, 0.1), np.arange(0.3, 0.8, 0.1), np.arange(0.5, 1.0, 0.1), np.arange(0.8, 1.1, 0.1), np.arange(1.5, 2.0, 0.1), np.arange(2.0, 2.5, 0.1), np.arange(4.0, 4.7 , 0.2), np.arange(5.0, 8.0 , 0.5) , np.arange(10.0, 22.0 , 2.0), np.arange(30.0, 42.0 , 2.0)]
#betas = np.array([[0.5, 0.7, 1.2], [1.0, 1.5, 2.0], [2, 3, 3.9], [3.2, 4.6, 6.1], [5, 7, 9], [8, 11.5, 15], [11, 15.5, 20], [15, 20, 25], [0, 35, 50], [0, 60, 80]])

alpha = 0.1
chi = 0.0
gamma = 5.0

ddir = 'data' #'/new_drho_data'#
model = '/model_tanh_implicit/vary_D' # '/model_hill'#

pdir = '../../'+ddir+model+'/K_{:d}/alpha_{:1.1f}_beta_{:1.1f}_chi_{:1.1f}_gamma_{:1.1f}/D_{:.1f}/run_{:d}/'.format(K, alpha, betas[0][0], chi, gamma, Ds[0], run)

datadir = pdir+'data/'

with open(pdir+'parameters.json') as jsonFile:
    parameters = json.load(jsonFile)

T         = parameters["T"]        # final time
dt_dump   = parameters["dt_dump"]
n_steps   = int(parameters["n_steps"])  # number of time steps
n_dump = int(T/dt_dump); dn_dump = int(n_steps/n_dump)
K         = parameters["K"]        # elastic constant, sets diffusion lengthscale of S with Gamma0
mx        = np.int32(parameters["mx"])
my        = np.int32(parameters["my"])
mu        = np.float32(parameters["mu"])
dx        = np.float32(parameters["dx"])
dy        = np.float32(parameters["dy"])
#setup a meshgrid
tol = 0.001

x   = np.linspace(0+tol, mx*dx-tol, mx)
y   = np.linspace(0+tol, mx*dx-tol, my)
xv, yv  = np.meshgrid(x,y, indexing='ij')


times = np.arange(0, n_dump, 1)*dt_dump

print(n_dump, T, mx, dx)
```

```python
ki=0

nr, nc = (1,1)
fig, ax = plt.subplots(figsize=(5*nc,5*nr), nrows=nr, ncols=nc)
ax.set_xscale('log')
ax.set_yscale('log')
ax.plot(Ds, np.power(Ds, 5/6), linestyle='dashed')

ax.set_xlabel('D', size=18)
ax.set_ylabel(r'$\beta_\mathrm{crit}$', size=18)

for di, D in enumerate(Ds):
    n_defect = np.zeros(len(betas[di]))
    for b, beta  in enumerate(betas[di]):
        n=9
        pdir = '../../'+ddir+model+'/K_{:d}/alpha_{:1.1f}_beta_{:1.1f}_chi_{:1.1f}_gamma_{:1.1f}/D_{:.1f}/run_{:d}/'.format(K, alpha, beta, chi, gamma, D, run)
        Qxx = np.loadtxt(pdir+'data/Qxx.csv.{:d}'.format(n), delimiter=',')
        Qxy = np.loadtxt(pdir+'data/Qxy.csv.{:d}'.format(n), delimiter=',')
        S = 2*np.sqrt(Qxx**2+Qxy**2)
    
        mask = S<0.3*np.max(S)
        l_mask = label(mask)
        regions = regionprops(l_mask)
        n_defect[b] = len(regions)
        if n_defect[b]==3 or n_defect[b]==4:
            marker = 'gd'
        elif n_defect[b]==0 or n_defect[b]==2:
            marker = 'ro'
        else:
            print(n_defect[b])
            marker='rp'
        ax.plot(D, beta, marker)
    sign = np.where(n_defect>2, 1, 0)
    beta_crit = ( betas[di][np.nonzero(sign)[0][0]] + betas[di][np.nonzero(sign)[0][0]-1] )/2
    ax.plot(D, beta_crit, 'ks')
plt.show()
```

```python
nr, nc = (2,3)
fig, ax = plt.subplots(figsize=(5*nc,5*nr), nrows=nr, ncols=nc)
fig.tight_layout()

Ds1 = [0.1, 1, 10]
beta_crits1 = [0.3, 1.2, 9.0]

for b, D in enumerate(Ds1):
    beta = beta_crits1[b]
    n=9
    pdir = '../../'+ddir+model+'/K_{:d}/alpha_{:1.1f}_beta_{:1.2f}_chi_{:1.1f}_gamma_{:1.1f}/D_{:.1f}/run_{:d}/'.format(K, alpha, beta, chi, gamma, D, run)
    Qxx = np.loadtxt(pdir+'data/Qxx.csv.{:d}'.format(n), delimiter=',')
    Qxy = np.loadtxt(pdir+'data/Qxy.csv.{:d}'.format(n), delimiter=',')
    S = 2*np.sqrt(Qxx**2+Qxy**2)

    p_factor = 3
    theta = np.arctan2(Qxy, Qxx)/2
    Snx    = np.cos(theta)
    Sny    = np.sin(theta)
    
    plot1 = [ax[b//nc, b%nc].pcolormesh(xv, yv, S, cmap='viridis', vmin=np.min(S), vmax=np.max(S)), ax[b//nc, b%nc].quiver(xv[::p_factor, ::p_factor], yv[::p_factor, ::p_factor], Snx[::p_factor, ::p_factor], Sny[::p_factor, ::p_factor], color='b', pivot='middle', headlength=0, headaxislength=0, scale=0.5, scale_units='xy')]
    ax[b//nc, b%nc].set_xticks([])
    ax[b//nc, b%nc].set_yticks([])
    fig.colorbar(plot1[0], ax=ax[b//nc, b%nc], shrink=0.75)
    ax[b//nc, b%nc].axis('equal')
    ax[b//nc, b%nc].axis('off')
    ax[b//nc, b%nc].set_title(r'$D={:.1f}$, $\beta$={:.1f}'.format(D, beta))

defect_loc = np.zeros((len(Ds)-1, 4, 2))

for di, D in enumerate(Ds[:-1]):
    beta = beta_crits[di]
    print(D, beta)
    n=9
    if beta==1.2 or beta==5.3:
        n=19
    pdir = '../../'+ddir+model+'/K_{:d}/alpha_{:1.1f}_beta_{:1.2f}_chi_{:1.1f}_gamma_{:1.1f}/D_{:.1f}/run_{:d}/'.format(K, alpha, beta, chi, gamma, D, run)
    Qxx = np.loadtxt(pdir+'data/Qxx.csv.{:d}'.format(n), delimiter=',')
    Qxy = np.loadtxt(pdir+'data/Qxy.csv.{:d}'.format(n), delimiter=',')
    S = 2*np.sqrt(Qxx**2+Qxy**2)

    mask = S<0.2
    l_mask = label(mask)
    regions = regionprops(l_mask)
    i=0
    for region in regions:
        minx, miny, maxx, maxy = region.bbox
        defect_loc[di, i] = np.array([(minx + maxx )/2,(miny + maxy)/2])
        i+=1
dx=0.5
dmm = defect_loc[:, 3, 1] - defect_loc[:, 0, 1]
ax[1,0].scatter(Ds[:-1], dmm * dx)
ax[1,0].set_xscale('log')
ax[1,0].set_ylabel(r'$d_{--}$', size=20)
ax[1,0].set_xlabel(r'$D$', size=20)

dpp =(defect_loc[:, 2, 1] - defect_loc[:,1, 1])*(np.sum(np.where(defect_loc[:,:,1]==0, 0, 1), axis=1)-3)
ax[1,1].scatter(Ds[:-1], dpp * dx)
ax[1,1].set_xscale('log')
ax[1,1].set_ylabel(r'$d_{++}$', size=20)
ax[1,1].set_xlabel(r'$D$', size=20)

dpm_1 = defect_loc[:, 1, 1] - defect_loc[:, 0, 1]
ax[1,2].scatter(Ds[:-1], dpm_1 * dx)
ax[1,2].set_xscale('log')
ax[1,2].set_ylabel(r'$d_{+-}$', size=20)
ax[1,2].set_xlabel(r'$D$', size=20)

plt.show()
```
