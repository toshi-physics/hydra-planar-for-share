```python
import numpy as np
from matplotlib import pyplot as plt
import json, os
from scipy.special import kn
#plt.style.use("./tplot.mplstyle")
```

```python
run = 1
beta  = 15.0
alpha = 0.1
chi = 0.0
gamma = 5.0
D=6.0
K=3

#mxs = np.array([210, 300, 200])
mxs = np.array([210, 100])
#dxs = np.array([1.0, 0.5, 0.5])
dxs = np.array([1.0, 1.0])

ddir = 'data' #'/new_drho_data'#
model = '/model_tanh_implicit_no_friction/strain_decay_expt' # '/model_hill'#

pdir = '../../'+ddir+model+'/mx_{:d}_D_{:1.1f}_K_{:d}/alpha_{:1.1f}_beta_{:1.4f}_chi_{:1.1f}_gamma_{:1.1f}/run_{:d}/'.format(mxs[0], D, K, alpha, beta, chi, gamma, run)

datadir = pdir+'data/'

with open(pdir+'parameters.json') as jsonFile:
    parameters = json.load(jsonFile)

T         = parameters["T"]        # final time
dt_dump   = parameters["dt_dump"]
n_steps   = int(parameters["n_steps"])  # number of time steps
n_dump = int(T/dt_dump); dn_dump = int(n_steps/n_dump)
K         = parameters["K"]        # elastic constant, sets diffusion lengthscale of S with Gamma0
mu        = np.float32(parameters["mu"])
lambd        = np.float32(parameters["lambd"])
dx        = np.float32(parameters["dx"])
dy        = np.float32(parameters["dy"])
tauc      = np.float32(parameters["tauc"])

times = np.arange(0, n_dump, 1)*dt_dump

print(n_dump, T, dx)
```

```python
def pade(r):
    return r*np.sqrt((0.34+(0.07*r**2))/(1+(0.41*r**2)+(0.07*r**4)))

n=19
midline_strains=[]
markers=['x', 'd']
linestyles=['dashed', 'dashdot']

r0s = np.zeros_like(mxs)
for i, m in enumerate(mxs):
    dx = dxs[i]
    
    datadir = '../../'+ddir+model+'/mx_{:d}_D_{:1.1f}_K_{:d}/alpha_{:1.1f}_beta_{:1.4f}_chi_{:1.1f}_gamma_{:1.1f}/run_{:d}/data/'.format(m, D, K, alpha, beta, chi, gamma, run)

    Qxx = np.loadtxt(datadir+'Qxx.csv.{:d}'.format(n), delimiter=',')
    Qxy = np.loadtxt(datadir+'Qxy.csv.{:d}'.format(n), delimiter=',')
    S = 2*np.sqrt(Qxx**2+Qxy**2)[m//2:, m//2]
    
    e_iso = np.loadtxt(datadir+'eiso.csv.{:d}'.format(n), delimiter=',')

    midline_strains.append(e_iso[m//2:, m//2])
    
    r0s[i] = dx*np.where(midline_strains[-1]<5e-3)[0][0]
    
    xx = np.arange(dx-0.01, (m//2)*dx, dx)
    
    plt.scatter(xx, midline_strains[-1], label=r'L={:.0f}'.format(m*dx), marker=markers[i])

    #plt.plot(xx[:8], 0.07-alpha*np.log(xx[:8])/2.5, color='k', linestyle='dotted')

    lq = np.sqrt(K); M=2.5

    S_dash = (S[1:]-S[:-1])/dx

    integral = np.sum(S/xx)-np.cumsum(S/xx)

    yy =  -alpha*(np.log(xx[-1])*2 - S[-1] + S - 2*integral*dx - 2*np.log(xx))/(2*M)

    plt.plot(xx, yy, linestyle=linestyles[i])
    
plt.ylabel('Strain profile +1')
plt.xlabel('r')
plt.xscale('log')
plt.legend()
```
