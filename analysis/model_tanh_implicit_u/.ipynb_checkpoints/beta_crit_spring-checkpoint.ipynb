```python
import numpy as np
from matplotlib import pyplot as plt
import json, os
from skimage.measure import label, regionprops
```

```python
K = np.array([1, 3, 5])
D = np.array([0.1, 0.3, 1, 3, 5, 8, 10, 20, 30])
beta_crit = np.array([[0.25, 0.45, 1.1, 2.5, 4.15, 6.5, 7.9, 15.5, 23],[0.65, 0.8, 1.8, 4.2, 6.5, 10, 12, 23, 35],[1.1, 1.7, 2.85, 6.5, 8.5, 13, 15, 28.9, 43]])
markers = np.array(['x', 'd', 's'])
cmap = plt.get_cmap('tab10')
colors = np.array([cmap(i) for i in np.linspace(0, 1, len(D))])

R=50
tauc =100
alpha=0.1
et = 0.5
xi =1.0
M=2.5
gamma=5
lambd_xi = np.sqrt(M/xi)
lambd = np.sqrt(D*tauc)
Sbar = alpha*tauc/(xi*et*(lambd**2 - lambd_xi**2))

yy = K[:, np.newaxis]/(16*gamma*Sbar[np.newaxis, :]**2)
xx = beta_crit**2

for k in np.arange(len(K)):
    plt.scatter(xx[k], yy[k], c=np.arange(1, len(D)+1, 1), cmap='Set1', marker=markers[k], label=r'K={:d}'.format(K[k]))

x = np.arange(0.1, 1000, 100)
plt.plot(x, x, color='k', linestyle='dotted')
cbar=plt.colorbar()
cbar.set_ticklabels(D)
cbar.ax.set_xlabel('D', rotation=0, size=12)

plt.xlabel(r'$K\dfrac{\xi^2\epsilon_t^2(\lambda^2-\lambda_\xi^2)^2}{16 \gamma \alpha^2\tau_c^2}$')
plt.ylabel(r'$\beta_c^2$')

plt.xscale('log')
plt.yscale('log')

leg=plt.legend(frameon=False)
```

```python
K = np.array([1, 3, 5])
D = np.array([0.1, 0.3, 1, 3, 5, 8, 10, 20, 30])
beta_crit = np.array([[0.25, 0.45, 1.1, 2.5, 4.15, 6.5, 7.9, 15.5, 23],[0.65, 0.8, 1.8, 4.2, 6.5, 10, 12, 23, 35],[1.1, 1.7, 2.85, 6.5, 8.5, 13, 15, 28.9, 43]])
markers = np.array(['x', 'd', 's'])
cmap = plt.get_cmap('tab10')
colors = np.array([cmap(i) for i in np.linspace(0, 1, len(D))])

R=50
tauc =100
alpha=0.1
et = 0.5
xi =1.0
M=2.5
gamma=5
lq = np.sqrt(K)
lambd_xi = np.sqrt(M/xi)
lambd = np.sqrt(D*tauc)
Sbar = alpha*tauc*np.pi/(8*et*M*lambd[np.newaxis,:])

yy = K[:, np.newaxis]*np.pi/(Sbar**2 * gamma*lq*(13*np.pi*lq/48))
xx = beta_crit**2

for k in np.arange(len(K)):
    plt.scatter(xx[k], yy[k], c=np.arange(1, len(D)+1, 1), cmap='Set1', marker=markers[k], label=r'K={:d}'.format(K[k]))

x = np.arange(0.1, 1000, 100)
plt.plot(x, x, color='k', linestyle='dotted')
cbar=plt.colorbar()
cbar.set_ticklabels(D)
cbar.ax.set_xlabel('D', rotation=0, size=12)

plt.xlabel(r'$K\dfrac{\xi^2\epsilon_t^2(\lambda^2-\lambda_\xi^2)^2}{16 \gamma \alpha^2\tau_c^2}$')
plt.ylabel(r'$\beta_c^2$')

plt.xscale('log')
plt.yscale('log')

leg=plt.legend(frameon=False)
```
